# Sigmoid 

```python
Image('./image/sigmoid_function.png')
```

$$z = \dfrac{1}{1 + e^{-t}}  ,  \dfrac{\partial z}{\partial t} = z(1 - z)  ,  t = W \cdot x $$

####  $x$

```python
x = np.array([[-1],[-2]])
x.shape 
```

#### $W$

```python
W = np.random.randn(1,2) # rand는 0~1, randn 은 표준정규분포
W.shape
```

#### 정방향, z는 W의 행 개수와 동일한 크기의 벡터 

```python
z = 1.0/(1.0+np.exp(-np.dot(W,x))) 
```

#### 역방향: x를 위한 로컬 그래디언트

```python
dx = np.dot(W.T, z*(1-z))
```

```python
dx
```

#### 역방향: W를 위한 로컬 그래디언트, outer로 원래 W 행렬 사이즈 복원

```python
dW = np.outer(z*(1-z), x)
```

```python
W
```

```python
dW
```

### Activation function이자 Output unit인 Sigmoid Function

```python
def sigmoid(x): 
    return 1.0 / (1.0 + np.exp(-x))
```

### Cost fuction 

이터레이션의 총 cost를 출력하기 위한 sum_of_error(batch 4개의 코스트 합치기)와 <br>
각 training set의 error를 미분하고, sigmoid function의 local gradient까지 계산한 결과를 간단하게 p-y로 정리하여<br>
backward function에서 hidden layer로 곧장 gradient를 넘긴다.

$\dfrac{\partial E}{\partial W} = \dfrac{\partial E}{\partial O}\dfrac{\partial O}{\partial Z}\dfrac{\partial Z}{\partial W}$

$ o = \dfrac{1}{1 + e^{-z}},$ output unit

* $E$는 Error
* $O$는 output layer의 결과(sigmoid 집어 넣은 후 결과)
* $Z$는 sigmoid에 집어넣는 z

```python
def cost(p):
    error = -((y * np.log(p) + (1-y) * np.log(1 - p)))
    sum_of_error = np.sum(error)/4.0
    dEdo = p - y # 왜 이렇게 하는지는 아래 링크 참고 바람
    return dEdo , sum_of_error
```

https://www.ics.uci.edu/~pjsadows/notes.pdf

### Forward path 

```python
def forward(x):
    h = sigmoid(np.dot(model['W1'], x)+1)
    logp = np.dot(model['W2'], h)+1
    p = sigmoid(logp)
    
    return p, h
```

### Backward path 

```python
def backward(h,e):
    dW2 = np.dot(h, e.T).ravel()
    dh = np.outer(e, model['W2'])
    dh = np.multiply(dh,1-dh)
    dW1 = np.dot(dh.T, x.T)
    return {'W1':dW1, 'W2':dW2}
```

```python
def train(learning_rate, iteration):
    
    for i in range(iteration):
        
        p, h = forward(x)
        e, error = cost(p)
        gradient = backward(h,e)
        for k,v in model.iteritems():
                model[k] -= learning_rate * gradient[k]

        if i % 1000 == 0:
            print("cost : ", error)
```

```python
train(0.5,20000)
```

# ReLu 

히든 레이어의 activation function으로 ReLu를 사용. Output unit은 여전히 sigmoid

bias term은 상수 1로 고정

http://stackoverflow.com/questions/2480650/role-of-bias-in-neural-networks

```python
model = {}
model['W1'] = np.random.rand(H,D) / np.sqrt(D) # Xavier Initialization
model['W2'] = np.random.rand(H) / np.sqrt(H)
```

```python
def lforward(x):
    # 2 layer의 NN
    # Output layer는 sigmoid(binary)
    h = np.dot(model['W1'],x)+1
    h[h<0] = 0 # ReLU nonlinearity
    logp = np.dot(model['W2'], h)+1
    p = sigmoid(logp)
    return p, h 
```

```python
def lbackward(hidden,loss):
    """ backward pass. (eph is array of intermediate hidden states) """
    dW2 = np.dot(hidden, loss.T).ravel()
    dh = np.outer(loss, model['W2'])
    dh[hidden.T <= 0] = 0 # backpro prelu
    dW1 = np.dot(dh.T, x.T)
    return {'W1':dW1, 'W2':dW2}
```

```python
def ltrain(learning_rate, iteration):
    
    for i in range(iteration):
        
        p, h = lforward(x)
        e, error = cost(p)
        gradient = lbackward(h,e)
        for k,v in model.iteritems():
                model[k] -= learning_rate * gradient[k]

        if i % 1000 == 0:
            print("cost : ", error)
```

```python
ltrain(0.5,20000)
```
