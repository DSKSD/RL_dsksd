## Mrkov Property 

"The future is independent of the past given the present"

```python
Image('./image/mdp_0.jpg')
```

t+1의 상태 S_t+1은 오직 전 스텝의 상태인 S_t에만 영향을 받는다.<br> P(S_t+1|S_t) = P(S_t+1|S_1,....,S_t) <br>
* state S_t는 그 전 모든 history에 관한 정보를 모두 포함한다.
* state가 한번 알려지면, 그 history는 아마 사라질 것이다?
* i.e. The state is a sufficient statistic of the future

## State Transition Matrix 

추이행렬(상태 전이 행렬)

P_ss' = P[S_t+1 = s' | S_t = s] # 상태 s에서 s'로 갈 확률
<br>
각 row의 합은 1이 되야 함

## Markov Process 

마코프 프로세스는 memoryless random process이다. 즉 랜덤한 스테이트들의 시퀀스.. <br>
A Markov Process(or Markov Chain) is a tuple (S,P)
* S is a (finite) set of states (유한 개의 상태공간)
* P is a state transition probability matrix(P는 추이행렬)
* P_ss' = P[S_t+1 = s' | S_t = s]

## Return 

```python
Image('./image/mdp_4.jpg')
```

이 리턴을 최대화 하는 것이 우리의 목표이며 감마는 0~1의 값으로 discount factor이다. (현재의 보상 대비 다음 스텝의 보상이 얼마만큼의 가치를 가질 것이냐) 현재의 보상이 미래의 보상보다 중요할 것이다.<br>
* 0에 가까울수록 근시안적(당장의 보상 중시)
* 1에 가까울수록 미래지향적(딜레이된 보상 고려)

### Why discount? 

우리는 완벽한 environment와 model을 설계할 수 없기 때문에, 예를 들어 '정확히 10 step 뒤에 reward가 100만큼 생길거야!' 를 확신할 수 없기 때문에..?

* 수학적으로 편리
* 마코프 프로세스가 순환적이여서 infinite를 리턴하는 것을 피하기 위해
* 정확히 나타낼 수 없는 미래에 대한 불확실성
* 만약 보상이 재무적이라면, 즉각적인 보상이 딜레이 된 보상보다 더 많은 가치를 담고 있다.
* 인간과 동물의 행동은 즉각적인 보상을 더 좋아하는 경향이 있다.
* 때때로 감마를 1로 주고 undiscounted Markov reward processes를 쓰는 것도 가능하다.
